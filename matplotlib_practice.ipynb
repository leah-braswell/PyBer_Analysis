```python
#import dependencies
import matplotlib.pyplot as plt
```

```python
#Set the x-axis to a list of strings for each month.
x_axis = ["Jan,", 'Feb', 'Mar', 'April', 'May', 'June', 'July', 'Aug', 'Sept', 'Oct', 'Nov', 'Dec']

#set the y-axis to a list of floats as the toatl fare in US dollars accumulated for each month.
y_axis = [10.2, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
#Create the plot
plt.plot(x_axis, y_axis)
```

```python
#Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
#Create the plot with ax.plt() the long way
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt() the quick way
ax = plt.axes()
ax.plot(x_axis, y_axis, color = 'green', linewidth = 2, marker  = 'd')
ax.legend("Boston")
ax.set_xlabel("Date")
ax.set_ylabel('Fare($)')
ax.set_title('PyBer Fare by Month')
ax.set_ylim(0, 45)
ax.grid()
```

```python
#Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
 #create the chart using plt.plot(x,y,'o') method
plt.plot(x_axis, y_axis, 'o')
```

```python
#create the chart using plt.scatter() method
plt.scatter(y_axis, x_axis, c = 'red', label = 'Chicago')
plt.legend()
plt.title("PyBer Fare by Month")
plt.ylabel('Date')
plt.xlabel('Fare($)')
plt.xlim(0, 50)
plt.gca().invert_yaxis()
```

```python
#create a bubble chart by changing the size of the marker
plt.scatter(x_axis, y_axis, s=y_axis)
```

```python
#change the size of all bubbles by iterating through the values in an axis and multiplying 
y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*3)
```

```python
#use the new y_axis list to set the size parameter
plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
#use list comprehension to multiply each data point in the y-axis
plt.scatter(x_axis, y_axis, s=[i * 3 for i in y_axis])
```

```python
#use obj oriented to create a scatter plot
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis)
```

```python
#create bubble by changing the size of the dot based on y-axis values
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis, s = y_axis)
```

```python
#create a bubble chart with fun stuff
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, c = 'skyblue', alpha = 0.2, edgecolors = 'black', linewidth = 2, s = [i*5 for i in y_axis], label = 'Boston')
ax.invert_yaxis()
ax.legend()
ax.set_xlim(0, 50)
ax.set_title('PyBer Fare by Month')
ax.set_xlabel('Fare($)')
ax.set_ylabel('Date')
```

```python
#create a pie chart with months as labels MATLAB
plt.pie(y_axis, labels = x_axis)
plt.show()
```

```python
#increase size using figsize
plt.subplots(figsize = (8,8))
#assign colors for each month with a list
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
#explode by adding a "distance from the center of the pie" for each value
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.pie(y_axis, 
        explode = explode_values, 
        labels = x_axis, 
        colors = colors, 
        autopct = '%1.1f%%')
```

```python
#create the chart using the obj. oriented method
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = [0, 0, .2, 0, 0, 0, .3, 0, 0, 0, 0, 0]
fig, ax = plt.subplots(figsize = (8,8))
ax.pie(y_axis, 
       explode = explode_values, 
       labels = x_axis, 
       colors = colors,
       autopct = '%1.1f%%', 
       shadow = True, 
       startangle = 90, 
       counterclock = False)
```
